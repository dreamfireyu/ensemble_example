# Weighted Average Ensemble Operator

```python
import towhee
towhee.glob['path']('./test.jpeg')\
    .image_decode['path','img']()\
    .image_embedding.timm['img','emb1'](model_name='resnet50')\
    .image_embedding.timm['img','emb2'](model_name='resnet101')\
    .image_embedding.timm['img','emb3'](model_name='resnet152')\
    .ensemble.weighted_average_ensemble[('emb1','emb2','emb3'),'emb'](weights=(0.3,0.3,0.4))\
    .select['img','emb']()\
    .show()
```

# Weighted Concat Ensemble Operator

```python
import towhee
towhee.glob['path']('./test.jpeg')\
    .image_decode['path','img']()\
    .image_embedding.timm['img','emb1'](model_name='resnet50')\
    .image_embedding.timm['img','emb2'](model_name='resnet101')\
    .image_embedding.timm['img','emb3'](model_name='resnet152')\
    .ensemble.weighted_concat_ensemble[('emb1','emb2','emb3'),'emb'](weights=(0.3,0.3,0.4))\
    .select['img','emb']()\
    .show()
```
